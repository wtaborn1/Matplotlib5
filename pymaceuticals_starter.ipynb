```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
drug_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
mouse_drug = pd.merge(mouse_data, drug_data, how="left", on=["Mouse ID", "Mouse ID"])


#pd.set_option('display.max_rows', None)
print(mouse_drug)


# Display the data table for preview


#########################

#mouse_data.head(50)
#drug_data.head(50)

         

```

```python
td =mouse_drug.groupby(["Drug", "Timepoint"]).count()[["Tumor Volume (mm3)","Metastatic Sites"]]


survival_count = pd.DataFrame(td).reset_index()
survival_count.head(45)
                   
                   
```

```python
plt.plot(survival_count.loc[survival_count["Drug"] == "Capomulin"]["Timepoint"],survival_count.loc[survival_count["Drug"] == "Capomulin"]["Tumor Volume (mm3)"])
plt.plot(survival_count.loc[survival_count["Drug"] == "Infubinol"]["Timepoint"],survival_count.loc[survival_count["Drug"] == "Infubinol"]["Tumor Volume (mm3)"])
plt.plot(survival_count.loc[survival_count["Drug"] == "Ketapril"]["Timepoint"],survival_count.loc[survival_count["Drug"] == "Ketapril"]["Tumor Volume (mm3)"])
plt.plot(survival_count.loc[survival_count["Drug"] == "Placebo"]["Timepoint"],survival_count.loc[survival_count["Drug"] == "Placebo"]["Tumor Volume (mm3)"])
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_count = survival_count.reset_index()
survival_count["Percentage"] = survival_count["Metastatic Sites"].astype("float64")/25*100
#survival_count_integer = pd.DataFrame(survival_count,dtype = "int32")
#survival_percentage = survival_count_integer / 25 * 100
print (survival_count)

# Preview the Data Frame
```

```python
plt.plot(survival_count.loc[survival_count["Drug"] == "Capomulin"]["Timepoint"],survival_count.loc[survival_count["Drug"] == "Capomulin"]["Percentage"])
plt.plot(survival_count.loc[survival_count["Drug"] == "Infubinol"]["Timepoint"],survival_count.loc[survival_count["Drug"] == "Infubinol"]["Percentage"])
plt.plot(survival_count.loc[survival_count["Drug"] == "Ketapril"]["Timepoint"],survival_count.loc[survival_count["Drug"] == "Ketapril"]["Percentage"])
plt.plot(survival_count.loc[survival_count["Drug"] == "Placebo"]["Timepoint"],survival_count.loc[survival_count["Drug"] == "Placebo"]["Percentage"])
```
